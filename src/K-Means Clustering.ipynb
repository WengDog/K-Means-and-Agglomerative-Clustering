```python
from copy import deepcopy
import numpy as np
import pandas as pd
import random
from sklearn import datasets

# Ambil data iris dari sklearn
iris = datasets.load_iris()
irisData = iris.data

# Tentukan centroid
centroid_1 = irisData[0]
centroid_2 = irisData[1]
centroid_3 = irisData[2]

# Gabungkan centroid dalam listOfCentroid
listOfCentroid = []
listOfCentroid.append(centroid_1)
listOfCentroid.append(centroid_2)
listOfCentroid.append(centroid_3)

# listOfCentroid akan menghasilkan [array([5.1, 3.5, 1.4, 0.2]), array([4.9, 3. , 1.4, 0.2]), array([4.7, 3.2, 1.3, 0.2])]!
# Tidak bisa diakses, ubah lagi menggunakan np.array
listOfCentroid = np.array(listOfCentroid)


print("Centroid_1")
print(centroid_1, "\n")

print("Centroid_2")
print(centroid_2, "\n")

print("Centroid_3")
print(centroid_3, "\n")

print("List Of Centroid")
print(listOfCentroid, "\n")

# Create distance function using euclidean distance
def euclidean_dist(a, b, ax=1):
    # numpy.linalg.norm(a-b) have default orde 2, which equals to euclidean formula
    return np.linalg.norm(a - b, axis = ax)

# Simpan nilai centroid yang akan diupdate, inisialisasi dengan 0
listOfCentroid_old = np.zeros(listOfCentroid.shape)
print(listOfCentroid_old)

# Simpan centroid yang terdekat dengan titik
clusters = np.zeros(len(irisData))
print(clusters)

# Simpan error awal, iteration akan berhenti sampai error = 0
error = euclidean_dist(listOfCentroid,listOfCentroid_old,None)
print(error)

k = 3 #jumlah cluster
while error!=0:
    # Masukkan setiap titik pada cluster terdekat
    for i in range(len(irisData)):
        #cari setiap jarak terhadap centroid
        distances = euclidean_dist(irisData[i],listOfCentroid)
        print("distances pada i = " + str(i) + " : " + str(distances))
        
        #np.argmin akan mengambil indeks dengan value terkecil
        cluster = np.argmin(distances)
        print("clusters pada i = " + str(i) + " : " + str(cluster) + "\n")
        
        #masukkan dalam dalam cluster
        clusters[i] = cluster
        
    # Simpan centroid sebelumnya, berguna untuk perhitungan error 
    listOfCentroid_old = deepcopy(listOfCentroid)
    # Cari rataan untuk setiap titik pada cluster
    for i in range(k):
        points = [irisData[j] for j in range(len(irisData)) if clusters[j] == i]
        listOfCentroid[i] = np.mean(points, axis=0)
    error = euclidean_dist(listOfCentroid, listOfCentroid_old, None)

print("Centroid akhir : ")
print(listOfCentroid, "\n")

print("Hasil clustering : ")
print(clusters, "\n")

print("error : ")
print(error)
```
